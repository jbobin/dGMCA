```python
n_s = 2
n_obs = 2
t_samp = 10000
rho = 0.05

X,X0,A0,S0,N = Make_Experiment_GG(n_s=n_s,n_obs=n_obs,t_samp=t_samp,noise_level=40.0,dynamic=1,CondNumber=2,alpha=rho)
```

## Get the optimization landscape 

```python
def GetArrayTheta(theta=0):
    return np.array([np.cos(theta/180*np.pi),np.sin(theta/180*np.pi)])

A0[:,0]=GetArrayTheta(theta=35)
A0[:,1]=GetArrayTheta(theta=70)

X = A0@S0

print(np.shape(X))
```

```python
nb = 128
v_thetaX = np.linspace(0,40,nb)
v_thetaY = np.linspace(45,85,nb)
u = np.zeros((nb,nb))
v = np.zeros((nb,nb))
A = np.copy(A0)

for rx in range(nb):
    for ry in range(nb):
        A[:,0]=GetArrayTheta(theta=v_thetaX[rx])
        A[:,1]=GetArrayTheta(theta=v_thetaY[ry]+1e-3)
        S = (np.linalg.inv(A.T@A)@A.T)@X
        R = A@S
        v[rx,ry] = np.sum(abs(S))
        u[rx,ry] = np.linalg.norm(X - R)
```

```python
plt.imshow(np.log10(v),vmax=1e-15)
```
