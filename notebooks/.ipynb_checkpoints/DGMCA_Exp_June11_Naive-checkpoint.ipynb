```python
import scipy.io as sio

critA_omp = []
critA = []
Time_omp = []
Time = []

totalSize = 10000
vBS = np.array([12,15,25,50,100,250,500,1000,2500])

n_s = 5 # Number of sources
n_obs = 20 # Number of observations        
rho = 0.1
Ainit = np.random.randn(n_obs,n_s)
Ainit = Ainit/np.maximum(0,np.linalg.norm(Ainit,axis=0))
kend = 1

BS = 10
N_MC = 1
nmax = 250

critA = []
critA_FMeq = []
Time_FMeq = []
critA_FM = []
Time_FM = []
critA_rFMeq = []
Time_rFMeq = []
critA_rFM = []
Time_rFM = []
critA_EUCeq = []
Time_EUCeq = []
critA_EUC = []
Time_EUC = []
critA_ODL = []
Time_ODL = []
critA_Naive_rFM = []

for nmc in range(N_MC):

    X,X0,A0,S0,N = Make_Experiment_GG(n_s=n_s,n_obs=n_obs,t_samp=totalSize,noise_level=40.0,dynamic=1,CondNumber=3,alpha=rho)
    
    for r in range(len(vBS)):

        BS = vBS[r]
        
        print(nmc,BS)
        
        R = np.dot(X,X.T)
        D,V = lng.eig(R)
        Ainit = V[:,0:n_s]
    
        print(nmc,totalSize,BS)
        
        Res = gmca.GMCA(X,n=n_s,mints=1.,nmax=250,L0=1,UseP=1,verb=0,Init=0,SCOpt=0,Kmax=1)
        critA.append(EvalCriterion_eval(A0,S0,Res['mixmat'],S0)['ca_mean'])
        
        ## Naive approach
        
        numBlocks = np.floor(X.shape[1]/(BS*1.)).astype(int)
        u = np.random.permutation(X.shape[1])
        Ntot = np.max([1,np.int(X.shape[1]/BS)])*BS
        Xbatches = np.reshape(X[:,u[0:Ntot]],(X.shape[0],BS,-1))
        Aest = np.zeros((n_obs,n_s,Xbatches.shape[2]))
        print(np.shape(Xbatches))
        for qq in range(Xbatches.shape[2]):
            Aest[:,:,qq] = gmca.GMCA(Xbatches[:,:,qq],n=n_s,mints=1.,nmax=250,L0=1,UseP=1,verb=0,Init=0,SCOpt=0,Kmax=1)['mixmat']
        A_mean = mp.FrechetMean(Aest,np.ones((n_s,Xbatches.shape[2]))/Xbatches.shape[2])
        critA_Naive_rFM.append(EvalCriterion_eval(A0,S0,A_mean,S0)['ca_mean'])
        
        ## dGMCA
        
        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,FullyRandom=True,kend=kend,AggMeth='FM',Equal=True,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS)
        t2 = time.time()
        Time_FMeq.append(t2-t1)
        critA_FMeq.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])
        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,FullyRandom=True,kend=kend,AggMeth='FM',Equal=False,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS)
        t2 = time.time()
        Time_FM.append(t2-t1)
        critA_FM.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])
        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,FullyRandom=True,kend=kend,AggMeth='rFM',Equal=True,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS)
        t2 = time.time()
        Time_rFMeq.append(t2-t1)
        critA_rFMeq.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])
        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,FullyRandom=True,kend=kend,AggMeth='rFM',Equal=False,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS)
        t2 = time.time()
        Time_rFM.append(t2-t1)
        critA_rFM.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])
        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,FullyRandom=True,kend=kend,AggMeth='Euclidean',Equal=True,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS)
        t2 = time.time()
        critA_EUCeq.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])
        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,FullyRandom=True,kend=kend,AggMeth='Euclidean',Equal=False,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS)
        t2 = time.time()
        Time_rFM.append(t2-t1)
        critA_EUC.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])
        t1 = time.time()
        dico = MiniBatchDictionaryLearning(n_components=n_s, alpha=0.05, n_iter=10000,batch_size=BS,fit_algorithm='cd')
        V = dico.fit(X.T).components_        
        t2 = time.time()
        Time_ODL.append(t2-t1)
        critA_ODL.append(EvalCriterion_eval(A0,S0,V.T,S0)['ca_mean'])
        
        sio.savemat("DGMCA_BatchSize_ns5_rho01_June11_Naive.mat",mdict={"critA_Naive_rFM":critA_Naive_rFM,"Time_ODL":Time_ODL,"critA_ODL":critA_ODL,"nmc":nmc,"Time":Time,"critA":critA,"Time_FMeq":Time_FMeq,"critA_FMeq":critA_FMeq,"Time_FM":Time_FM,"critA_FM":critA_FM,"Time_rFMeq":Time_rFMeq,"critA_rFMeq":critA_rFMeq,"Time_rFM":Time_rFM,"critA_rFM":critA_rFM,"Time_EUCeq":Time_EUCeq,"critA_EUCeq":critA_EUCeq,"Time_EUC":Time_EUC,"critA_EUC":critA_EUC})
```
