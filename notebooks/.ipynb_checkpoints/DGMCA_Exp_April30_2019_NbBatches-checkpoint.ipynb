```python
import sys
sys.path.append("../src/")
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from utils_dgmca import divisorGenerator
from utils_dgmca import Make_Experiment_GG
from misc_dgmca import *
import copy as cp
from tqdm import tqdm
import DGMCA_omp as dgo
import time
import scipy.io as sio
```

```python
import scipy.io as sio

critA_omp = []
critA = []
Time_omp = []
Time = []

totalSize = 1000000
n_s = 5 # Number of sources
n_obs = 10 # Number of observations        
rho = 0.1
Ainit = np.random.randn(n_obs,n_s)
Ainit = Ainit/np.maximum(0,np.linalg.norm(Ainit,axis=0))

vBS = np.array([100,500,1000,5000,10000,50000,100000])
N_MC = 10
BS = 10
nmax = 1000
FullyRandom = False

critA_FMeq = []
Time_FMeq = []
critA_FM = []
Time_FM = []
critA_rFMeq = []
Time_rFMeq = []
critA_rFM = []
Time_rFM = []
critA_EUCeq = []
Time_EUCeq = []
critA_EUC = []
Time_EUC = []

for nmc in range(N_MC):

    Xin,X0,A0,S0,N = Make_Experiment_GG(n_s=n_s,n_obs=n_obs,t_samp=totalSize,noise_level=40.0,dynamic=1,CondNumber=1,alpha=rho)
        
    for r in range(len(vBS)):
                
        ts = vBS[r]
        print(nmc,ts)
        
        X = Xin[:,0:ts]
        
        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,kend=1.,AggMeth='FM',Equal=False,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6,FullyRandom=FullyRandom, subBlockSize=BS,alphaEstOpt=1,alpha_exp=2.)
        t2 = time.time()
        Time_FM.append(t2-t1)
        critA_FM.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])

        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,kend=1.,AggMeth='rFM',Equal=False,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6,FullyRandom=FullyRandom, subBlockSize=BS,alphaEstOpt=1,alpha_exp=2.)
        t2 = time.time()
        Time_rFM.append(t2-t1)
        critA_rFM.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])
        
        sio.savemat("DGMCA_BS_10_NbBatches_rho01_2_Nov19.mat",mdict={"Time":Time,"critA":critA,"Time_FMeq":Time_FMeq,"critA_FMeq":critA_FMeq,"Time_FM":Time_FM,"critA_FM":critA_FM,"Time_rFMeq":Time_rFMeq,"critA_rFMeq":critA_rFMeq,"Time_rFM":Time_rFM,"critA_rFM":critA_rFM,"Time_EUCeq":Time_EUCeq,"critA_EUCeq":critA_EUCeq,"Time_EUC":Time_EUC,"critA_EUC":critA_EUC})

vBS = np.array([1000,5000,10000,50000,100000,500000])
N_MC = 10
BS = 100
nmax = 1000

critA_FMeq = []
Time_FMeq = []
critA_FM = []
Time_FM = []
critA_rFMeq = []
Time_rFMeq = []
critA_rFM = []
Time_rFM = []
critA_EUCeq = []
Time_EUCeq = []
critA_EUC = []
Time_EUC = []

for nmc in range(N_MC):

    Xin,X0,A0,S0,N = Make_Experiment_GG(n_s=n_s,n_obs=n_obs,t_samp=totalSize,noise_level=40.0,dynamic=1,CondNumber=1,alpha=rho)
        
    for r in range(len(vBS)):
                
        ts = vBS[r]
        print(nmc,ts)
        
        X = Xin[:,0:ts]
        
        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,kend=1.,AggMeth='FM',Equal=False,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6,FullyRandom=FullyRandom, subBlockSize=BS,alphaEstOpt=1,alpha_exp=2.)
        t2 = time.time()
        Time_FM.append(t2-t1)
        critA_FM.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])

        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,kend=1.,AggMeth='rFM',Equal=False,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6,FullyRandom=FullyRandom, subBlockSize=BS,alphaEstOpt=1,alpha_exp=2.)
        t2 = time.time()
        Time_rFM.append(t2-t1)
        critA_rFM.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])
        
        sio.savemat("DGMCA_BS_100_NbBatches_rho01_2_Nov19.mat",mdict={"Time":Time,"critA":critA,"Time_FMeq":Time_FMeq,"critA_FMeq":critA_FMeq,"Time_FM":Time_FM,"critA_FM":critA_FM,"Time_rFMeq":Time_rFMeq,"critA_rFMeq":critA_rFMeq,"Time_rFM":Time_rFM,"critA_rFM":critA_rFM,"Time_EUCeq":Time_EUCeq,"critA_EUCeq":critA_EUCeq,"Time_EUC":Time_EUC,"critA_EUC":critA_EUC})

vBS = np.array([10000,50000,100000,500000,1000000])
N_MC = 10
BS = 1000
nmax = 1000

critA_FMeq = []
Time_FMeq = []
critA_FM = []
Time_FM = []
critA_rFMeq = []
Time_rFMeq = []
critA_rFM = []
Time_rFM = []
critA_EUCeq = []
Time_EUCeq = []
critA_EUC = []
Time_EUC = []

for nmc in range(N_MC):

    Xin,X0,A0,S0,N = Make_Experiment_GG(n_s=n_s,n_obs=n_obs,t_samp=totalSize,noise_level=40.0,dynamic=1,CondNumber=1,alpha=rho)
        
    for r in range(len(vBS)):
                
        ts = vBS[r]
        print(nmc,ts)
        
        X = Xin[:,0:ts]
        
        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,kend=1.,AggMeth='FM',Equal=False,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,FullyRandom=FullyRandom,tol=1e-6, subBlockSize=BS,alphaEstOpt=1,alpha_exp=2.)
        t2 = time.time()
        Time_FM.append(t2-t1)
        critA_FM.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])

        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,kend=1.,AggMeth='rFM',Equal=False,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,FullyRandom=FullyRandom,tol=1e-6, subBlockSize=BS,alphaEstOpt=1,alpha_exp=2.)
        t2 = time.time()
        Time_rFM.append(t2-t1)
        critA_rFM.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])
        
        sio.savemat("DGMCA_BS_1000_NbBatches_rho01_2_Nov19.mat",mdict={"Time":Time,"critA":critA,"Time_FMeq":Time_FMeq,"critA_FMeq":critA_FMeq,"Time_FM":Time_FM,"critA_FM":critA_FM,"Time_rFMeq":Time_rFMeq,"critA_rFMeq":critA_rFMeq,"Time_rFM":Time_rFM,"critA_rFM":critA_rFM,"Time_EUCeq":Time_EUCeq,"critA_EUCeq":critA_EUCeq,"Time_EUC":Time_EUC,"critA_EUC":critA_EUC})
```

```python
### Plotting the results

N_MC = 5
vBS = np.array([1000,5000,10000,50000,100000])
nval = len(vBS)
ntot = N_MC*nval
totalSize = 10000

data = sio.loadmat("DGMCA_BS_10_NbBatches_April30.mat")

Time_FMeq = data["Time_FMeq"]
Time_FM = data["Time_FM"]
Time_rFMeq = data["Time_rFMeq"]
Time_rFM = data["Time_rFM"]
Time_EUCeq = data["Time_EUCeq"]
Time_EUC = data["Time_EUC"]
critA_FMeq = data["critA_FMeq"]
critA_FM = data["critA_FM"]
critA_rFMeq = data["critA_rFMeq"]
critA_rFM = data["critA_rFM"]
critA_EUCeq = data["critA_EUCeq"]
critA_EUC = data["critA_EUC"]
Time = data["Time"]
critA = data["critA"]

CompTime = np.mean(Time)
Crit = np.mean(critA)
DCompTime_FMeq = np.mean(np.reshape(Time_FMeq[0:ntot],(N_MC,-1)),axis=0)
DCrit_FMeq = np.mean(np.reshape(critA_FMeq[0:ntot],(N_MC,-1)),axis=0)
DCompTime_FM = np.mean(np.reshape(Time_FM[0:ntot],(N_MC,-1)),axis=0)
DCrit_FM = np.mean(np.reshape(critA_FM[0:ntot],(N_MC,-1)),axis=0)
DCompTime_EUCeq = np.mean(np.reshape(Time_EUCeq[0:ntot],(N_MC,-1)),axis=0)
DCrit_EUCeq = np.mean(np.reshape(critA_EUCeq[0:ntot],(N_MC,-1)),axis=0)
DCompTime_EUC = np.mean(np.reshape(Time_EUC[0:ntot],(N_MC,-1)),axis=0)
DCrit_EUC = np.mean(np.reshape(critA_EUC[0:ntot],(N_MC,-1)),axis=0)
DCompTime_rFMeq = np.mean(np.reshape(Time_rFMeq[0:ntot],(N_MC,-1)),axis=0)
DCrit_rFMeq = np.mean(np.reshape(critA_rFMeq[0:ntot],(N_MC,-1)),axis=0)
DCompTime_rFM = np.mean(np.reshape(Time_rFM[0:ntot],(N_MC,-1)),axis=0)
DCrit_rFM = np.mean(np.reshape(critA_rFM[0:ntot],(N_MC,-1)),axis=0)

fsize = 16
vcol = ['mediumseagreen','crimson','steelblue','darkmagenta','burlywood','khaki','lightblue','darkseagreen','deepskyblue','forestgreen','gold','indianred','midnightblue','olive','orangered','orchid','red','steelblue']
font = {'family' : 'normal',
            'weight' : 'bold',
            'size'   : fsize}
plt.rc('font', **font)
plt.rcParams["figure.figsize"] = (15,10)

plt.loglog(vBS,DCrit_FMeq,'--s',alpha=0.7,lw=4,markersize=10,color=vcol[0])
#plt.semilogy(vBS,DCrit_FM)
plt.loglog(vBS,DCrit_rFMeq,'d--',lw=4,alpha=0.7,markersize=10,color=vcol[1])
#plt.semilogy(vBS,DCrit_rFM,'d')
plt.loglog(vBS,DCrit_EUCeq,'o--',lw=4,alpha=0.7,markersize=10,color=vcol[2])
#plt.semilogy(vBS,DCrit_EUC,'o')
plt.loglog(vBS,Crit*np.ones_like(DCrit_FMeq),'--p',alpha=0.7,lw=4,markersize=10,color=vcol[3])

plt.legend(["FMeq","rFMeq","EUCeq","NoBatch"])

plt.ylabel("Mixing matrix criterion",fontsize = fsize+2,weight='bold')
plt.xlabel("Number of samples",fontsize = fsize+2,weight='bold')
plt.title('BS = 10',fontsize = fsize+2,weight='bold')
plt.savefig('NoWeighting_BS10_Cd1_ns5_nobs20')

plt.close()

### with the weighting

plt.loglog(vBS,DCrit_FM,'--s',alpha=0.7,lw=4,markersize=10,color=vcol[0])
#plt.semilogy(vBS,DCrit_FM)
plt.loglog(vBS,DCrit_rFM,'d--',lw=4,alpha=0.7,markersize=10,color=vcol[1])
#plt.semilogy(vBS,DCrit_rFM,'d')
plt.loglog(vBS,DCrit_EUC,'o--',lw=4,alpha=0.7,markersize=10,color=vcol[2])
#plt.semilogy(vBS,DCrit_EUC,'o')
plt.loglog(vBS,Crit*np.ones_like(DCrit_FMeq),'--p',alpha=0.7,lw=4,markersize=10,color=vcol[3])

plt.legend(["FMeq","rFMeq","EUCeq","NoBatch"])

plt.ylabel("Mixing matrix criterion",fontsize = fsize+2,weight='bold')
plt.xlabel("Number of samples",fontsize = fsize+2,weight='bold')
plt.title('BS = 10',fontsize = fsize+2,weight='bold')
plt.savefig('NoWeighting_BS10_Cd1_ns5_nobs20')
```
