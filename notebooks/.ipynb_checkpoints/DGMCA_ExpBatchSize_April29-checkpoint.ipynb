```python
totalSize = 10000
n_s = 5 # Number of sources
BlockSize = n_s
n_obs = 10 # Number of observations
rho = 0.1 # Generalized Gaussian generation parameter
kend = 2.
nmax= 1000
X,X0,A0,S0,N = Make_Experiment_GG(n_s=n_s,n_obs=n_obs,t_samp=totalSize,noise_level=40.0,dynamic=1,CondNumber=1,alpha=rho)
Ainit = np.random.randn(n_obs,n_s)
Ainit = Ainit/np.maximum(0,np.linalg.norm(Ainit,axis=0))


BSize = np.array([5,10,20,50,100,200,500,1000,5000])

rFM = []
FM = []
wrFM = []
wFM = []
Euc = []
wEuc = []

for BS in BSize:
    print("Batch size = ",BS)
    out_omp250,FMw250,th250,Oca250,Osad250,Ocam250,Osadm250 = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,kend=kend,q_f=0.1,AggMeth='FM',Equal=True,nmax=nmax,BlockSize = BlockSize,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS, SCOpt=1,alphaEstOpt=1,alpha_exp=2.,A0=A0)
    FM.append(Ocam250[-1])
    out_omp250,FMw250,th250,Oca250,Osad250,Ocam250,Osadm250 = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,kend=kend,q_f=0.1,AggMeth='FM',Equal=False,nmax=nmax,BlockSize = BlockSize,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS, SCOpt=1,alphaEstOpt=1,alpha_exp=2.,A0=A0)
    wFM.append(Ocam250[-1])
    out_omp250,FMw250,th250,Oca250,Osad250,Ocam250,Osadm250 = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,kend=kend,q_f=0.1,AggMeth='rFM',Equal=True,nmax=nmax,BlockSize = BlockSize,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS, SCOpt=1,alphaEstOpt=1,alpha_exp=2.,A0=A0)
    rFM.append(Ocam250[-1])
    out_omp250,FMw250,th250,Oca250,Osad250,Ocam250,Osadm250 = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,kend=kend,q_f=0.1,AggMeth='rFM',Equal=False,nmax=nmax,BlockSize = BlockSize,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS, SCOpt=1,alphaEstOpt=1,alpha_exp=2.,A0=A0)
    wrFM.append(Ocam250[-1])
    out_omp250,FMw250,th250,Oca250,Osad250,Ocam250,Osadm250 = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,kend=kend,q_f=0.1,AggMeth='Euclidean',Equal=True,nmax=nmax,BlockSize = BlockSize,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS, SCOpt=1,alphaEstOpt=1,alpha_exp=2.,A0=A0)
    Euc.append(Ocam250[-1])
    out_omp250,FMw250,th250,Oca250,Osad250,Ocam250,Osadm250 = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,kend=kend,q_f=0.1,AggMeth='Euclidean',Equal=False,nmax=nmax,BlockSize = BlockSize,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS, SCOpt=1,alphaEstOpt=1,alpha_exp=2.,A0=A0)
    wEuc.append(Ocam250[-1])
```
