```python
import sys
sys.path.append('../src/')
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from utils_dgmca import divisorGenerator
from utils_dgmca import Make_Experiment_GG
from misc_dgmca import *
from DGMCA import DGMCA as dgmca
from DGMCA_omp import *
from GMCA import Core_GMCA as gmca
import copy as cp
from tqdm import tqdm
import time
import scipy.io as sio
import DGMCA_omp as dgo
import DGMCA as dg
import GMCA as gmca
```

```python
### SUPER SPARSE CASE WHERE DGMCA WORKS BETTER

totalSize = 10000
n_s = 5 # Number of sources
n_obs = 20 # Number of observations
rho = 0.1 # Generalized Gaussian generation parameter
BlockSize =  n_s
nmax = 1000
kend = 1.

# We should do MC

X,X0,A0,S0,N = Make_Experiment_GG(n_s=n_s,n_obs=n_obs,t_samp=totalSize,noise_level=40.0,dynamic=1,CondNumber=3,alpha=rho)

Ainit = np.random.randn(n_obs,n_s)

Agmca = gmca.GMCA(X,n=n_s,mints=1.,nmax=250,L0=1,UseP=1,verb=0,Init=0,SCOpt=0,Kmax=1)["mixmat"]
cA = EvalCriterion_eval(A0,S0,Agmca,S0)['ca_mean']
```

```python
FR = False

nmax=100

BS = 100
out_omp100,FMw100,th1000,Oca100,Osad100,Ocam100,Osadm100 = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,FullyRandom=FR,kend=kend,q_f=0.1,AggMeth='FM',Equal=True,nmax=nmax,BlockSize = BlockSize,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS, A0=A0)
BS = 1000
out_omp1000,FMw1000,th1000,Oca1000,Osad1000,Ocam1000,Osadm1000 = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,FullyRandom=FR,kend=kend,q_f=0.1,AggMeth='FM',Equal=True,nmax=nmax,BlockSize = BlockSize,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS, A0=A0)
BS = 10
out_omp10,FMw10,th10,Oca10,Osad10,Ocam10,Osadm10 = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,FullyRandom=FR,kend=kend,q_f=0.1,AggMeth='FM',Equal=True,nmax=nmax,BlockSize = BlockSize,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS, A0=A0)
BS = 100
out_omp100,FMw100,th1000,rOca100,rOsad100,rOcam100,rOsadm100 = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,FullyRandom=FR,kend=kend,q_f=0.1,AggMeth='rFM',Equal=True,nmax=nmax,BlockSize = BlockSize,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS, A0=A0)
BS = 1000
out_omp1000,rFMw1000,rth1000,rOca1000,rOsad1000,rOcam1000,rOsadm1000 = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,FullyRandom=FR,kend=kend,q_f=0.1,AggMeth='rFM',Equal=True,nmax=nmax,BlockSize = BlockSize,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS, A0=A0)
BS = 10
out_omp10,FMw10,th10,rOca10,rOsad10,rOcam10,rOsadm10 = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,FullyRandom=FR,kend=kend,q_f=0.1,AggMeth='rFM',Equal=True,nmax=nmax,BlockSize = BlockSize,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS,A0=A0)
print("DONE")

sio.savemat("CaHist_FM_rho0_1_cond3_June17.mat",mdict={"gmca":cA,"Oca100":Oca100,"Oca1000":Oca1000,"Oca10":Oca10,"rOca100":rOca100,"rOca1000":rOca1000,"rOca10":rOcam10,"Ocam100":Ocam100,"Ocam1000":Ocam1000,"Ocam10":Ocam10,"rOcam100":rOcam100,"rOcam1000":rOcam1000,"rOcam10":rOcam10})
```
