```python
import sys
sys.path.append("../src/")
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from utils_dgmca import divisorGenerator
from utils_dgmca import Make_Experiment_GG
from misc_dgmca import *
import copy as cp
#from tqdm import tqdm
import DGMCA_omp as dgo
import time
import scipy.io as sio
```

## BS = 10

```python
import scipy.io as sio

critA_omp = []
critA = []
Time_omp = []
Time = []

BS = 10

vBS = BS*np.array([5,50,100,500,1000,2000,3000,5000])
n_s = 5 # Number of sources
n_obs = 20 # Number of observations        
rho = 0.05
Ainit = np.random.randn(n_obs,n_s)
Ainit = Ainit/np.maximum(0,np.linalg.norm(Ainit,axis=0))
kend = 1

N_MC = 25
nmax = 1000

critA_FMeq = []
Time_FMeq = []
critA_FM = []
Time_FM = []
critA_rFMeq = []
Time_rFMeq = []
critA_rFM = []
Time_rFM = []
critA_EUCeq = []
Time_EUCeq = []
critA_EUC = []
Time_EUC = []

for nmc in range(N_MC):
    
    for r in range(len(vBS)):
        
        totalSize = vBS[r]
        
        print(BS,nmc,totalSize)

        X,X0,A0,S0,N = Make_Experiment_GG(n_s=n_s,n_obs=n_obs,t_samp=totalSize,noise_level=40.0,dynamic=1,CondNumber=2,alpha=rho)
    

        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,kend=kend,AggMeth='FM',Equal=True,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS, SCOpt=1,alphaEstOpt=1,alpha_exp=2.)
        t2 = time.time()
        Time_FMeq.append(t2-t1)
        critA_FMeq.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])
        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,kend=kend,AggMeth='FM',Equal=False,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS, SCOpt=1,alphaEstOpt=1,alpha_exp=2.)
        t2 = time.time()
        Time_FM.append(t2-t1)
        critA_FM.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])
        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,kend=kend,AggMeth='rFM',Equal=True,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS, SCOpt=1,alphaEstOpt=1,alpha_exp=2.)
        t2 = time.time()
        Time_rFMeq.append(t2-t1)
        critA_rFMeq.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])
        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,kend=kend,AggMeth='rFM',Equal=False,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS, SCOpt=1,alphaEstOpt=1,alpha_exp=2.)
        t2 = time.time()
        Time_rFM.append(t2-t1)
        critA_rFM.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])
        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,kend=kend,AggMeth='Euclidean',Equal=True,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS, SCOpt=1,alphaEstOpt=1,alpha_exp=2.)
        t2 = time.time()
        Time_EUCeq.append(t2-t1)
        critA_EUCeq.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])
        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,kend=kend,AggMeth='Euclidean',Equal=False,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS, SCOpt=1,alphaEstOpt=1,alpha_exp=2.)
        t2 = time.time()
        Time_EUC.append(t2-t1)
        critA_EUC.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])
        
        sio.savemat("DGMCA_Cd2_TotSize_ns5_BS10_25NMC_May3.mat",mdict={"Time":Time,"critA":critA,"Time_FMeq":Time_FMeq,"critA_FMeq":critA_FMeq,"Time_FM":Time_FM,"critA_FM":critA_FM,"Time_rFMeq":Time_rFMeq,"critA_rFMeq":critA_rFMeq,"Time_rFM":Time_rFM,"critA_rFM":critA_rFM,"Time_EUCeq":Time_EUCeq,"critA_EUCeq":critA_EUCeq,"Time_EUC":Time_EUC,"critA_EUC":critA_EUC})
```
