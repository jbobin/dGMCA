```python
### Plotting the results

N_MC = 25
totalSize = 10000
vBS = np.array([1.,2,3,5,7,10,15])
nval = len(vBS)
ntot = N_MC*nval

data = sio.loadmat("DGMCA_BS10_ns5_CdNumber_alpha01_25MC_June18_FR.mat")

Time_FMeq = data["Time_FMeq"]
Time_FM = data["Time_FM"]
Time_rFMeq = data["Time_rFMeq"]
Time_rFM = data["Time_rFM"]
Time_EUCeq = data["Time_EUCeq"]
Time_EUC = data["Time_EUC"]
critA_FMeq = data["critA_FMeq"]
critA_FM = data["critA_FM"]
critA_rFMeq = data["critA_rFMeq"]
critA_rFM = data["critA_rFM"]
critA_EUCeq = data["critA_EUCeq"]
critA_EUC = data["critA_EUC"]
Time = data["Time"]
critA = data["critA"]

GetMean = False
Weighted = True
fname = 'CdNumber_BS10'
xlab = 'Condition number'
ylab = 'Mixing matrix criterion'
pmax = 75
pmin = 25


Crit25 = np.percentile(critA,pmin)
DCrit_FMeq25 = np.percentile(np.reshape(critA_FMeq[0:ntot],(N_MC,-1)),pmin,axis=0)
DCrit_FM25 = np.percentile(np.reshape(critA_FM[0:ntot],(N_MC,-1)),pmin,axis=0)
DCrit_EUCeq25 = np.percentile(np.reshape(critA_EUCeq[0:ntot],(N_MC,-1)),pmin,axis=0)
DCrit_EUC25 = np.percentile(np.reshape(critA_EUC[0:ntot],(N_MC,-1)),pmin,axis=0)
DCrit_rFMeq25 = np.percentile(np.reshape(critA_rFMeq[0:ntot],(N_MC,-1)),pmin,axis=0)
DCrit_rFM25 = np.percentile(np.reshape(critA_rFM[0:ntot],(N_MC,-1)),pmin,axis=0)

Crit75 = np.percentile(critA,pmax)
DCrit_FMeq75 = np.percentile(np.reshape(critA_FMeq[0:ntot],(N_MC,-1)),pmax,axis=0)
DCrit_FM75 = np.percentile(np.reshape(critA_FM[0:ntot],(N_MC,-1)),pmax,axis=0)
DCrit_EUCeq75 = np.percentile(np.reshape(critA_EUCeq[0:ntot],(N_MC,-1)),pmax,axis=0)
DCrit_EUC75 = np.percentile(np.reshape(critA_EUC[0:ntot],(N_MC,-1)),pmax,axis=0)
DCrit_rFMeq75 = np.percentile(np.reshape(critA_rFMeq[0:ntot],(N_MC,-1)),pmax,axis=0)
DCrit_rFM75 = np.percentile(np.reshape(critA_rFM[0:ntot],(N_MC,-1)),pmax,axis=0)


if GetMean:

    Crit = np.mean(critA)
    DCrit_FMeq = np.mean(np.reshape(critA_FMeq[0:ntot],(N_MC,-1)),axis=0)
    DCrit_FM = np.mean(np.reshape(critA_FM[0:ntot],(N_MC,-1)),axis=0)
    DCrit_EUCeq = np.mean(np.reshape(critA_EUCeq[0:ntot],(N_MC,-1)),axis=0)
    DCrit_EUC = np.mean(np.reshape(critA_EUC[0:ntot],(N_MC,-1)),axis=0)
    DCrit_rFMeq = np.mean(np.reshape(critA_rFMeq[0:ntot],(N_MC,-1)),axis=0)
    DCrit_rFM = np.mean(np.reshape(critA_rFM[0:ntot],(N_MC,-1)),axis=0)
    
else:
    
    Crit = np.median(critA)
    DCrit_FMeq = np.median(np.reshape(critA_FMeq[0:ntot],(N_MC,-1)),axis=0)
    DCrit_FM = np.median(np.reshape(critA_FM[0:ntot],(N_MC,-1)),axis=0)
    DCrit_EUCeq = np.median(np.reshape(critA_EUCeq[0:ntot],(N_MC,-1)),axis=0)
    DCrit_EUC = np.median(np.reshape(critA_EUC[0:ntot],(N_MC,-1)),axis=0)
    DCrit_rFMeq = np.median(np.reshape(critA_rFMeq[0:ntot],(N_MC,-1)),axis=0)
    DCrit_rFM = np.median(np.reshape(critA_rFM[0:ntot],(N_MC,-1)),axis=0)

fsize = 16
vcol = ['mediumseagreen','crimson','steelblue','darkmagenta','burlywood','khaki','lightblue','darkseagreen','deepskyblue','forestgreen','gold','indianred','midnightblue','olive','orangered','orchid','red','steelblue']
font = {'family' : 'normal',
            'weight' : 'bold',
            'size'   : fsize}
plt.rc('font', **font)
plt.rcParams["figure.figsize"] = (15,10)
fig = plt.figure()
ax = plt.gca()

if Weighted:
    
    plt.loglog(vBS,DCrit_FM,'--s',alpha=0.7,lw=4,markersize=10,color=vcol[0])
    ax.fill_between(vBS, DCrit_FM25, DCrit_FM75, interpolate=True, alpha=0.1,color=vcol[0])
    plt.loglog(vBS,DCrit_rFM,'d--',lw=4,alpha=0.7,markersize=10,color=vcol[1])
    ax.fill_between(vBS, DCrit_rFM25, DCrit_rFM75, interpolate=True, alpha=0.1,color=vcol[1])
    plt.loglog(vBS,DCrit_EUC,'o--',lw=4,alpha=0.7,markersize=10,color=vcol[2])
    ax.fill_between(vBS, DCrit_EUC25, DCrit_EUC75, interpolate=True, alpha=0.1,color=vcol[2])
    plt.loglog(vBS,Crit*np.ones_like(DCrit_FM),'--p',alpha=0.7,lw=4,markersize=10,color=vcol[3])
    ax.fill_between(vBS, Crit25*np.ones_like(DCrit_FMeq), Crit75*np.ones_like(DCrit_FMeq), interpolate=True, alpha=0.1,color=vcol[3])

    plt.legend(["Frechet Mean","robust Frechet Mean","Euclidean","Full batch size"])

    plt.ylabel(ylab,fontsize = fsize+2,weight='bold')
    plt.xlabel(xlab,fontsize = fsize+2,weight='bold')
    plt.savefig('Weighted_'+fname)

else:

    plt.loglog(vBS,DCrit_FMeq,'--s',alpha=0.7,lw=4,markersize=10,color=vcol[0])
    ax.fill_between(vBS, DCrit_FMeq25, DCrit_FMeq75, interpolate=True, alpha=0.1,color=vcol[0])
    plt.loglog(vBS,DCrit_rFMeq,'d--',lw=4,alpha=0.7,markersize=10,color=vcol[1])
    ax.fill_between(vBS, DCrit_rFMeq25, DCrit_rFMeq75, interpolate=True, alpha=0.1,color=vcol[1])
    plt.loglog(vBS,DCrit_EUCeq,'o--',lw=4,alpha=0.7,markersize=10,color=vcol[2])
    ax.fill_between(vBS, DCrit_EUCeq25, DCrit_EUCeq75, interpolate=True, alpha=0.1,color=vcol[2])
    plt.loglog(vBS,Crit*np.ones_like(DCrit_FMeq),'--p',alpha=0.7,lw=4,markersize=10,color=vcol[3])
    ax.fill_between(vBS, Crit25*np.ones_like(DCrit_FMeq), Crit75*np.ones_like(DCrit_FMeq), interpolate=True, alpha=0.1,color=vcol[3])

    plt.legend(["Frechet Mean","robust Frechet Mean","Euclidean","Full batch size"])

    plt.ylabel(ylab,fontsize = fsize+2,weight='bold')
    plt.xlabel(xlab,fontsize = fsize+2,weight='bold')
    plt.savefig('NotWeighted_'+fname)
```
