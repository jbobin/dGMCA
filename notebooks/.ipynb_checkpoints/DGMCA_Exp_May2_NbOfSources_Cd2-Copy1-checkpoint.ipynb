```python
import sys
sys.path.append("../src/")
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from utils_dgmca import divisorGenerator
from utils_dgmca import Make_Experiment_GG
from misc_dgmca import *
import copy as cp
#from tqdm import tqdm
import DGMCA_omp as dgo
import time
import scipy.io as sio
from sklearn.decomposition import MiniBatchDictionaryLearning
import GMCA as gmca
```

```python
import scipy.io as sio

FR = False

critA_omp = []
critA = []
Time_omp = []
Time = []

totalSize = 10000
vBS = np.array([12,15,25,50,100,250,500,1000,2500])

n_s = 5 # Number of sources
n_obs = 20 # Number of observations        
rho = 0.1
Ainit = np.random.randn(n_obs,n_s)
Ainit = Ainit/np.maximum(0,np.linalg.norm(Ainit,axis=0))
kend = 1

BS = 10
N_MC = 10
nmax = 1000

critA = []
critA_FMeq = []
Time_FMeq = []
critA_FM = []
Time_FM = []
critA_rFMeq = []
Time_rFMeq = []
critA_rFM = []
Time_rFM = []
critA_EUCeq = []
Time_EUCeq = []
critA_EUC = []
Time_EUC = []
critA_ODL = []
Time_ODL = []

for nmc in range(N_MC):
    
    print("MC #",nmc)
    
    Ainit = np.random.randn(n_obs,n_s)
    Ainit = Ainit/np.linalg.norm(Ainit,axis=0)

    X,X0,A0,S0,N = Make_Experiment_GG(n_s=n_s,n_obs=n_obs,t_samp=totalSize,noise_level=40.0,dynamic=1,CondNumber=3,alpha=rho)
    
    for r in range(len(vBS)):

        BS = vBS[r]
        
        Res = gmca.GMCA(X,n=n_s,mints=1.,nmax=250,L0=1,UseP=1,verb=0,Init=0,SCOpt=0,Kmax=1,AInit = Ainit)
        critA.append(EvalCriterion_eval(A0,S0,Res['mixmat'],S0)['ca_mean'])
        
        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,FullyRandom=FR,kend=kend,AggMeth='FM',Equal=True,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS)
        t2 = time.time()
        Time_FMeq.append(t2-t1)
        critA_FMeq.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])
        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,FullyRandom=FR,kend=kend,AggMeth='FM',Equal=False,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS)
        t2 = time.time()
        Time_FM.append(t2-t1)
        critA_FM.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])
        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,FullyRandom=FR,kend=kend,AggMeth='rFM',Equal=True,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS)
        t2 = time.time()
        Time_rFMeq.append(t2-t1)
        critA_rFMeq.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])
        t1 = time.time()
        out_omp,FMw,th = dgo.Core_DGMCA(X=X,n=n_s,A=Ainit,FullyRandom=FR,kend=kend,AggMeth='rFM',Equal=False,q_f=0.1,nmax=nmax,L0=1,verb=0,Kmax=0.5,tol=1e-6, subBlockSize=BS)
        t2 = time.time()
        Time_rFM.append(t2-t1)
        critA_rFM.append(EvalCriterion_eval(A0,S0,out_omp,S0)['ca_mean'])
        t1 = time.time()
        t1 = time.time()
        dico = MiniBatchDictionaryLearning(n_components=n_s, alpha=0.05, n_iter=10000,batch_size=BS,fit_algorithm='cd')
        V = dico.fit(X.T).components_        
        t2 = time.time()
        Time_ODL.append(t2-t1)
        critA_ODL.append(EvalCriterion_eval(A0,S0,V.T,S0)['ca_mean'])
        
        sio.savemat("DGMCA_BatchSize_ns5_rho01_June3_NoFR.mat",mdict={"Time_ODL":Time_ODL,"critA_ODL":critA_ODL,"nmc":nmc,"Time":Time,"critA":critA,"Time_FMeq":Time_FMeq,"critA_FMeq":critA_FMeq,"Time_FM":Time_FM,"critA_FM":critA_FM,"Time_rFMeq":Time_rFMeq,"critA_rFMeq":critA_rFMeq,"Time_rFM":Time_rFM,"critA_rFM":critA_rFM,"Time_EUCeq":Time_EUCeq,"critA_EUCeq":critA_EUCeq,"Time_EUC":Time_EUC,"critA_EUC":critA_EUC})
```
