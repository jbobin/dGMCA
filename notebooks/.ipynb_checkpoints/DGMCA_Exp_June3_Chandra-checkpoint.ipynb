```python

Npix = 128
n_s = 3 # Number of sources
n_obs = 50 # Number of observations
J = 1

def GetChandraData(J=2,snr=40):

    import astropy.io.fits as pyf

    S0 = pyf.getdata("../../PoissonBSS/Sources.fits").reshape((4,128*128))
    S0 = np.dot(np.diag(1./np.linalg.norm(S0,axis=1)),S0)
    A0 = pyf.getdata("../../PoissonBSS/Spectra.fits")
    A0 = A0[125:175,[0,2,3]]
    S0 = S0[[0,2,3],:]
    A0 = A0/np.maximum(0,np.linalg.norm(A0,axis=0))

    for r in range(n_s):
        S0[r,:] = (S0[r,:] - np.min(S0[r,:]))
        S0[r,:] = S0[r,:]/np.linalg.norm(S0[r,:])

    X0 = np.dot(A0,S0) 
    N = np.random.randn(50,128*128)

    X = X0 + np.power(10.,-snr/20.)*np.linalg.norm(X0)/np.linalg.norm(N)*N

    Temp = X.reshape((n_obs,Npix,Npix))
    mw = ps.forward(Temp,J=J)
    Xw = np.reshape(mw[:,:,:,0:J],(n_obs,-1))

    return Xw,X,X0,A0,S0,N

```

```python
npix = 128
m = 50

def blockshaped(arr, nrows,ncols): ## Extracting non-overlapping patches
        u,h, w = arr.shape
        return (arr.reshape(u,h//nrows, nrows, -1, ncols)
                   .swapaxes(2,3)
                   .reshape(u,-1, nrows, ncols))
    
def Formatting_Data(X,BS=128,nscales = 2): 
    
    def blockshaped(arr, nrows,ncols): ## Extracting non-overlapping patches
        u,h, w = arr.shape
        return (arr.reshape(u,h//nrows, nrows, -1, ncols)
                   .swapaxes(2,3)
                   .reshape(u,-1, nrows, ncols))
    
    Xout = np.zeros((m,npix*npix))
    
    for scale in range(nscales):
        f = X[:,scale*npix*npix:(scale+1)*npix*npix]
        s = blockshaped(f, BS, BS)
        s = np.reshape(s,(m,-1,BS*BS)).swapaxes(0,2)
        #s =np.reshape(s,(BS*BS,-1))
        for r in range(s.shape[2]):
            Xout[r,scale*npix*npix:(scale+1)*npix*npix] = np.reshape(s[:,:,r],(-1,),order='F')
    
    return Xout
```
